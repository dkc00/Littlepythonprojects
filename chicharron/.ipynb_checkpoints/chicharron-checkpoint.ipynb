```python
# Comparison of chicharrón prices in Laureles, Medellín and their review ratings on Google
# Chicharrón restaurants were analyzed in the center and near Laureles. Chicharrón prices were taken from their actual menu.

import matplotlib.pyplot as plt
import csv
import numpy as np

#read csv file
file = open("laureleschicharron.csv")
```

```python
ratings = []
minprice = []
maxprice = []


#Load csv file 
with open("laureleschicharron.csv", newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    next(reader)  # the header line should not be read but only the values
    for row in reader:
        # Split list objects of csv file
        x = [i.split(";") for i in row]
        z = [j for i in x for j in i]     
        minprice.append(float(z[0]))   
        maxprice.append(float(z[1]))
        ratings.append(float(z[2])) 

'''
For further projects: Take care of not confusing data types. The minprice and maxprice has to be converted into a number,
before utilizing it in the plot. Due to this error, a plt.ylim(x,y) was not possible because the program read y values as strings.

'''
```

```python
#Plot
plt.style.use("seaborn-v0_8-bright")
plt.scatter(ratings, minprice, label="Minimal price")
plt.scatter(ratings, maxprice, label="Maximal price")
plt.title("Chicharrón prices in Laureles, Medellín")
plt.xlabel("Google Rating")
plt.ylabel("Price [COP]")
plt.xlim(4.0,5.0)
plt.ylim(10000,70000)
plt.legend()
plt.show()
```

```python
# Statistical description of my chicharrón results with numpy module

#convert to np arrays for more efficient code
minprice = np.array(minprice)
maxprice = np.array(maxprice)

# Calculation of the most important statistical parameters (Mean, Median, Varianz, Standard deviation)
mean_minprice = np.mean(minprice)
median_minprice = np.median(minprice)
variance_minprice = np.var(minprice)
stddev_minprice = np.std(minprice)

mean_maxprice = np.mean(maxprice)
median_maxprice = np.median(maxprice)
variance_maxprice = np.var(maxprice)
stddev_maxprice = np.std(maxprice)

# Print
print(f"Minimal Price, Mean: {mean_minprice}, Median: {median_minprice}, Variance: {variance_minprice}, Std. Deviation: {stddev_minprice}")
print(f"Maximal Price, Mean: {mean_maxprice}, Median: {median_maxprice}, Variance: {variance_maxprice}, Std, Deviation: {stddev_maxprice}")
```
