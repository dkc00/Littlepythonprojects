```python
# Comparison of chicharrón prices in Laureles, Medellín and their review ratings on Google

import matplotlib.pyplot as plt
import csv

#read csv file
file = open("laureleschicharron.csv")
```

```python
ratings = []
minprice = []
maxprice = []


#Load csv file 
with open("laureleschicharron.csv", newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    next(reader)  # the header line should not be read but only the values
    for row in reader:
        # Split list objects of csv file
        x = [i.split(";") for i in row]
        z = [j for i in x for j in i]
        ratings.append(float(z[2]))  
        minprice.append(z[0])     
        maxprice.append(z[1])
```

```python
#Plot
plt.style.use("seaborn-v0_8-bright")
plt.plot(ratings, minprice, label="Minimal price")
plt.plot(ratings, maxprice, label="Maximal price")
plt.title("Chicharrón prices in Laureles, Medellín")
plt.xlabel("Google Rating")
plt.ylabel("Price [COP]")
plt.xlim(4.0,5.0)

# plt.ylim(10000.00, 70000.00) Funktioniert nicht. Was läuft schief?
plt.legend()
plt.show()
```
